```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Objective function and its derivative
def objective(x):
    return (x + 3) ** 2

def derivative(x):
    return 2 * (x + 3)

# Gradient descent function
def gradient_descent(alpha, start, max_iter):
    x_values = [start]
    x = start
    for _ in range(max_iter):
        x -= alpha * derivative(x)  # Update x
        x_values.append(x)
    return x_values
```

```python
# Parameters for gradient descent
alpha = 0.1  # Learning rate
start = 2    # Starting point
max_iter = 30  # Maximum iterations
```

```python
# Generate x values from gradient descent
x_vals = gradient_descent(alpha, start, max_iter)
```

```python
# Plotting the function and gradient descent steps
x_range = np.linspace(-10, 5, 100)
plt.plot(x_range, objective(x_range), label=r'$y = (x + 3)^2$')
plt.plot(x_vals, objective(np.array(x_vals)), '.-', color='red', label="Gradient Descent Path")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Gradient Descent to Minimize $y = (x + 3)^2$")
plt.grid()
plt.show()
```
